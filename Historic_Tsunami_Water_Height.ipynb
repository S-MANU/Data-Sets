```python
df1 = df.loc[:,['Sr.no', 'Country', 'Year', 'Latitude','Longitude', 'EarthquakeNanMagnitude','MaximumNanWaterNanHeightNan(m)']]
df1.columns = ['Sr.no', 'Country', 'Year','Latitude', 'Longitude','Magnitude','Maximum Water Height(m)']
df1 = df1[df1['Maximum Water Height(m)']> 10]
df1
```

```python
df1=df1[['Sr.no','Country','Year','Magnitude','Maximum Water Height(m)']]
df1['Country_Year']=df['Country']+'_'+df['Year'].astype(str)
df1=df1.sort_values('Maximum Water Height(m)',ascending=False)
df1
```

```python
fig = px.bar(df1[0:30], y='Maximum Water Height(m)', x='Country_Year', title="Ranking of Historical Maximum Water Height")
fig.show()
```

##Location of Lituya Bay and the Epiccenter¶

```python
df2=df[['Latitude','Longitude']].dropna()
lat1=df[df['Sr.no']==651][['Latitude','Longitude']].iloc[0].tolist()
print('The epiccenter',lat1,'is the center of the map.')
print('Lituya Bay [58.643372,-137.8542578] is seen on the upper left of the map.')
eq_map = folium.Map(location=lat1,tiles='Stamen Terrain',zoom_start=8.0,min_zoom=2.0)

eq_map.add_child(plugins.HeatMap(df2))
eq_map
```
