##Project Summary

The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland. Explore and analyze the data to discover key findings pertaining to terrorist activities.

# Data Vizualization, Storytelling & Experimenting with charts

Total Number Of Attacks From Each Year.

```python
# Group by 'Year' and count occurrences
attacks = df.groupby('Year').size().reset_index(name='Attacks')

# Set 'Year' as the index
attacks.set_index('Year', inplace=True)

# Display the first few rows
print(attacks.head())
```

Number of attacks from each year, the above table shows head of the table.

Let's see the number of attacks with visualisation.

```python
# Set the style
sns.set(style="whitegrid")


# Plotting a bar graph
plt.figure(figsize=(15, 6))
sns.barplot(x=attacks.index, y='Attacks', data=attacks, color='black')

# Adding title and labels
plt.title("Timeline of Attacks", fontsize=15)
plt.xlabel("Years", fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)

# Rotate x-axis labels vertically
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

1. In the above figure we can see 2014 has the highest number of attacks.
2. In the above figure we can see 1971 has the lowest attacks.

Lets see the Number of Attcks on Every region.

```python
# Group by 'Year' and 'Region', then unstack the 'Region' column
reg = df.groupby(['Year', 'Region']).size().unstack(fill_value=0)

# Display the first few rows
print(reg.head())
```

The above table shows the number of attacks on each region.

```python
# Transpose the DataFrame for stacked area plot
reg_transposed = reg.T

# Plotting a stacked area plot
plt.figure(figsize=(20, 10))
reg_transposed.plot.area(stacked=True, alpha=0.5, colormap='viridis', figsize=(20, 10))

# Adding title and labels
plt.title("Region wise attacks", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Attacks", fontsize=20)

# Show the plot
plt.show()
```

In Eastern Europe, North America and South Asia we can see the high number of attacks.

Total Attacks on Each Region

```python
# Sum along the columns (axis=1) to get the total attacks for each region
tot_reg = reg.sum(axis=0).sort_values(ascending=False)

# Display the result
print(tot_reg)
```

```python
# Set the style
sns.set(style="whitegrid")

# Plotting a bar graph using Seaborn
plt.figure(figsize=(15, 6))
sns.barplot(x=tot_reg.index, y=tot_reg.values, palette="viridis")

# Adding title and labels
plt.title("Total Number of Attacks in each Region from 1970-2017", fontsize=15)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Number of Attacks", fontsize=12)

# Rotate x-axis labels if needed
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
