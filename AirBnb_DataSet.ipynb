## Lets dig deep with visualization

Lets find out how much percentage the neighbourhood hotels has the each type of room. so that we can see which type of rooms are higher and which type are lower etc.

```python
# Set Figure Size
plt.rcParams['figure.figsize'] = (8, 5)

# Create Count Plot
ax = sns.countplot(y='room_type', hue='neighbourhood_group', data=df, palette= "pastel")

# Calculate Total Number of Rooms
total = len(df['room_type'])

# Annotate Percentages on Bars
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width() / total)
    x = p.get_x() + p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y))

# Add Title and Labels
plt.title('Total number of rooms with percentage')
plt.xlabel('Rooms')
plt.xticks(rotation=90)
plt.ylabel('Room counts')

# Show the Plot
plt.show()
```

In the above figure we get a big picture of each city with their percentage of
private room , home and shared room. Here we can clearly see Manhattan and Brooklyn shares the most number of rooms.

## Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

Lets find out which type of rooms has the highest price.

```python
# Group the columns to plot
grouped_data = df.groupby('room_type')['price'].mean().reset_index()

# Create a bar plot using Seaborn
sns.barplot(x='room_type', y='price', data=grouped_data, width=0.3)
plt.title('Bar Plot Of Price Against Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')

# Show the plot
plt.show()
```

From above plot we can see that home has the highest price, then for private rooms and in the end shared rooms. Also there is a euge gap in price between home price and room price.

Lets find out how many nights people are staying in respective rooms.

```python
# Group the columns to plot
grouped_data = df.groupby('room_type')['minimum_nights'].mean().reset_index()

# Create a bar plot using Seaborn
sns.barplot(x='room_type', y='minimum_nights', data=grouped_data, width=0.3)
plt.title('Bar Plot of Minimum Nights by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Minimum Nights')

# Show the plot
plt.show()
```

Acoording to the above graph we can conclude that people more people stayed in the home, then shared room after that only private room.

Lets find out which type of rooms has the highest number of reviews

```python
# Group the columns to plot
grouped_data = df.groupby('room_type')['number_of_reviews'].mean().reset_index()

# Create a bar plot using Seaborn
sns.barplot(x='room_type', y='number_of_reviews', data=grouped_data,width=0.3)
plt.title('Bar Plot of Number Of Reviews by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Reviews')

# Show the plot
plt.show()
```

From the above graph almost equal number of reviews for private and home, least is for shared rooms.

Lets find out which type of rooms took highest number of hosts.

```python
# Group the columns to plot
grouped_data = df.groupby('room_type')['calculated_host_listings_count'].mean().reset_index()

# Create a bar plot using Seaborn
sns.barplot(x='room_type', y='calculated_host_listings_count', data=grouped_data, width=0.3)
plt.title('Bar Plot of Number Of Hosts by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Hosts')

# Show the plot
plt.show()
```

From the above figure we can see that home took the more hosts, then shared room and less was takes by private rooms.

Lets find which type rooms available most of the time in a year.

```python
# Group the columns to plot
grouped_data = df.groupby('room_type')['availability_365'].mean().reset_index()

# Create a bar plot using Seaborn
sns.barplot(x='room_type', y='availability_365', data=grouped_data, width=0.3)
plt.title('Bar Plot of Availability of Rooms by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Availability')

# Show the plot
plt.show()
```

From the above figure we can conclude that shared room availability is more in a year compare to home and private room types.

Lets Find The Busiest host In The Data

```python
busiest_host = df.groupby(['host_id','host_name','neighbourhood_group','room_type' ])['number_of_reviews'].max().reset_index()
busiest_host = busiest_host.sort_values(by ='number_of_reviews', ascending=False).head(10)
busiest_host
```

The above list are the top ten busiest hosts.

```python
hosts_name = busiest_host['host_name']
num_reviews = busiest_host['number_of_reviews']

plt.plot(hosts_name, num_reviews, marker='o', color='purple')
plt.xlabel('Hosts Name')
plt.ylabel('Number of Reviews')
plt.title('Line Plot of Number of Reviews by Hosts Name')
plt.show()
```

So Dona has the busiest host based on  reviews in this data. The above line graph shows top ten busiest hosts with respect to reviews.

Lets Find The Highest Charging Hosts

```python
High_Price = df.groupby(['host_id','host_name','neighbourhood_group','room_type'])['price'].max().reset_index()
High_Price = High_Price.sort_values(by = 'price', ascending=False).head(10)
High_Price
```

```python
hosts_name = High_Price['host_name']
high_price = High_Price['price']

plt.pie(high_price, labels=hosts_name, autopct='%1.1f%%', startangle=90, colors=['yellow', 'orange', 'green', 'red'])
plt.title('Pie Chart of Price by Hosts Name')
plt.show()
```

We can see highest price making hosts in above pie chart

Traffic Between The Areas

```python
traffic_areas = df.groupby(['neighbourhood_group', 'room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by ='minimum_nights',ascending=False).head(10)
traffic_areas
```

The above list are the maximum traffic areas in that Manhattan is the top.

```python
area_traffic = traffic_areas['room_type']
room_stayed = traffic_areas['minimum_nights']

plt.barh(area_traffic, room_stayed, color='orange', height=0.2)
plt.xlabel('Minimum Nights Stayed')
plt.ylabel('Room Type')
plt.title('Horizontal Bar Plot of Traffic Areas Based on Room Booked')
plt.show()
```

The above graph shows the most traffic room type. where home takes the more traffic.
